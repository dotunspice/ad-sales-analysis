Advertising Sales Analysis

This project analyzes how different advertising channels (TV, Radio, Newspaper) impact sales.

Objective

To identify which marketing channel contributes most to sales performance.

Key Insights
TV advertising has the strongest impact on sales
Radio has a moderate effect
Newspaper has minimal influence

Tools Used
Python
Pandas
Matplotlib

```python
import pandas as pd

df = pd.read_csv('/kaggle/input/datasets/dotunspice/ad-impact-on-sales-analysis/Advertising Budget and Sales.csv')
df.head()
```

```python
df.corr()
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['TV Ad Budget ($)'], df['Sales ($)'])
plt.xlabel('TV Ads')
plt.ylabel('Sales')
plt.title('TV Ads vs Sales')
plt.show()
```

```python
plt.scatter(df['Radio Ad Budget ($)'], df['Sales ($)'])
plt.xlabel('Radio Ads')
plt.ylabel('Sales')
plt.title('Radio Ads vs Sales')
plt.show()
```

```python
plt.scatter(df['Newspaper Ad Budget ($)'], df['Sales ($)'])
plt.xlabel('Newspaper Ads')
plt.ylabel('Sales')
plt.title('Newspaper Ads vs Sales')
plt.show()
```
